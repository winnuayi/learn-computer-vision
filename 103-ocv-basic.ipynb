Load image. Get image size. Render image in RGB format.

```python
# https://pyimagesearch.com/2018/07/19/opencv-tutorial-a-guide-to-learn-opencv/

import cv2
from matplotlib import pyplot as plt

# render image
def cv2_imshow(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)


# load image
image = cv2.imread("images/ciheul.jpg")

# image size
(h, w, d) = image.shape
print("width={}, height={}, depth={}".format(w, h, d))

cv2_imshow(image)
```

```python
# cari warna pixel pada x=50, y=100
(B, G, R) = image[100, 50]
print("R={}, G={}, B={}".format(R, G, B))
```

```python
# cropping gambar (region of interest)
roi = image[980:1200, 700:1200]
cv2_imshow(roi)
```

```python
resized = cv2.resize(image, (200, 200))
cv2_imshow(resized)
```

```python
r = 300.0 / w
dim = (300, int(h * r))
resized_2 = cv2.resize(image, dim)
cv2_imshow(resized_2)
```

```python
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
cv2_imshow(rotated)
```

```python
# blur gambar untuk mengurangi high-frequency noise
# sehingga algoritma lebih mudah untuk mendeteksi dan
# dan memahami konten aktual dari gambar
blurred = cv2.GaussianBlur(image, (11, 11), 0)
cv2_imshow(blurred)
```

```python
# gambar kotak merah
output = image.copy()
cv2.rectangle(output, (1100, 970), (1190, 1070), (0, 0, 255), 2)
cv2_imshow(output)
```

```python
# tambahkan text
output = image.copy()
cv2.putText(output, "Ciheul Technologies", (100, 100), cv2.FONT_HERSHEY_SIMPLEX,
            2.0, (0, 255, 0), 5)
cv2_imshow(output)
```
